# bar graph

```python
#3. Plot a bar graph of male vs female buyers participated in the sales.
```

```python
male=data['Buyer Gender'].value_counts()['Male']
female=data['Buyer Gender'].value_counts()['Female']

print('No.of males:',male)
print('No.of females',female)
```

```python
x=data['Buyer Gender'].value_counts()
```

```python
x
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
w=0.4
x=[1]
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]
fig = plt.figure(figsize =(8, 6))
plt.bar(bar1,male,w,label='Male')
plt.bar(bar2,female,w,label='Female')
plt.xlabel('Gender',fontsize=16)
plt.ylabel('No.of buyers ',fontsize=16)
plt.title('male vs female buyers participated in sales',fontsize=18)
plt.xticks(bar1+w/2,x)
plt.legend()
```

            * Therre are 5052 female buyers are in the list
            * There are 4948 male buyers are in the list
